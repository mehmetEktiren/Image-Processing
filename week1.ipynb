```python
m,n,p=im_1.shape
m,n,p
```

```python
plt.imshow(im_1)
plt.show()
```

```python
new_image=np.zeros((m,n),dtype=float)
new_image
```

```python
for i in range(m):
    for j in range (n):
        s=(im_1[i,j,0]+im_1[i,j,1]+im_1[i,j,2]/3)
        new_image[i,j]=s
```

```python
plt.imshow(new_image, cmap='gray')
plt.show()
plt.imsave('antep_2siyahbeyaz.jpeg', new_image,cmap='gray')
```
